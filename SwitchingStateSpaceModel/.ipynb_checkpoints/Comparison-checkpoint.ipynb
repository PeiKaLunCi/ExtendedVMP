```julia
using Plots, Random, DelimitedFiles, DataFrames
```

```julia
FLestimates = readdlm("FLestimates.txt")
HMCestimates = readdlm("HMCestimates.txt")
NUTSestimates = readdlm("NUTSestimates.txt")
;
```

```julia
Random.seed!(0);
#Generate data
T = 200

w1, w2 = 1, 10

y_data = [sqrt(1/w1)*randn()]
for t=2:50
    append!(y_data, y_data[end] + sqrt(1/w1)*randn())
end
for t=51:150
    append!(y_data, y_data[end] + sqrt(1/w2)*randn())
end
for t=151:T
    append!(y_data, y_data[end] + sqrt(1/w1)*randn())
end
```

```julia
gr()
l = @layout [a ; b; c; d]
mm=Plots.PlotMeasures.mm
p1 = plot(1:200, y_data, w=2, labels = "Observations", legend=(:bottomleft))
vspan!(p1,[1,50], color = :green, alpha = 0.2, labels = "Regime 1");
vspan!(p1,[50,150], color = :red, alpha = 0.2, labels = "Regime 2");
vspan!(p1,[150,T], color = :green, alpha = 0.2, labels = "");

p2 = plot(Gray.(FLestimates), link=:x, ylabel="E-VMP", top_margin=-20.0mm)
yticks!([5:10:15;], ["Regime 1", "Regime 2"])

p3 = plot(Gray.(HMCestimates), link=:x, ylabel="HMC", top_margin=-30.0mm)
yticks!([5:10:15;], ["Regime 1", "Regime 2"])

p4 = plot(Gray.(NUTSestimates), link=:x, xlabel="Time steps", ylabel="NUTS", top_margin=-30.0mm)
yticks!([5:10:15;], ["Regime 1", "Regime 2"])

p = plot(p1, p2, p3, p4, layout = l, size = (1000, 500))
```

```julia
png(p, "switching-state-space.png")
```
